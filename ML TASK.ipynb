## classification

```python
df=pd.read_csv('Movie_classification.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
median=df['Time_taken'].median()
df['Time_taken']=df['Time_taken'].fillna(median)
```

```python
df['Time_taken'].isnull().sum()
```

```python
df_gen=pd.get_dummies(df['Genre'],drop_first=True)
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
df=(x.drop(columns='Genre',axis=1))
```

```python
df=pd.concat([df,df_gen,y],axis=1)
```

```python
df.head()
```

```python
df['3D_available']=np.where(df['3D_available']=='YES',1,0)
```

```python
df.head()
```

```python
X=df.iloc[:,:-1].values
Y=df.iloc[:,-1].values
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import accuracy_score,classification_report,r2_score
model=DecisionTreeRegressor()
model=RandomForestClassifier()
```

```python
def classifier(alg,X_train, X_test, Y_train, Y_test):
    model=alg
    model.fit(X_train,Y_train)
    model.score(X_train,Y_train)
    predicted=model.predict(X_test)
    print(accuracy_score(predicted,Y_test))
    print(classification_report(predicted,Y_test))
    print(r2_score(predicted,Y_test))
```

```python
classifier(LogisticRegression(),X_train, X_test, Y_train, Y_test)
```

```python
classifier(RandomForestClassifier(),X_train, X_test, Y_train, Y_test)
```

```python
classifier(KNeighborsClassifier(),X_train, X_test, Y_train, Y_test)
```
